#### Reverse a string

```python
## Using built-in function

str = input("Enter a string")
print(str[::-1])
```

```python
##without built-in function

str1 = input("Enter a string")
final_string = ""
for i in str1:
    final_string = i + final_string
print(final_string)
    
```

#### Fibonacci Series Generator

```python
n = int(input("Enter the number:"))
a, b = 0, 1
for i in range(n):
    print(a, end=" ")
    a, b = b, a+b
```
