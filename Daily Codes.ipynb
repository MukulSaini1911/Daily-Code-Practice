#### Reverse a string

```python
## Using built-in function

str1 = input("Enter a string")
print(str1[::-1])
```

```python
##without built-in function

str1 = input("Enter a string")
final_string = ""
for i in str1:
    final_string = i + final_string
print(final_string)
```

#### Fibonacci Series Generator

```python
n = int(input("Enter the number:"))
a, b = 0, 1
for i in range(n):
    print(a, end=" ")
    a, b = b, a+b
```

#### Check Leap Year

```python
year = int(input("Enter the value of year:"))
if (year % 4 == 0 and year%100 != 0) or (year % 400 == 0):
    print("Given year is a leap Year")
else:
    print("Given year is not a leap year")                 
```

#### Function to check a String/Number is palindrome or not
#### String

```python
str1 = input("Enter the string:")
if str1.lower() == str1.lower()[::-1]:
    print("Palindrome")
else:
    print("Not a palindrome")
```

```python
## Using function
def palindrome_check(str1):
    return str1.lower() == str1.lower()[::-1]
print(palindrome_check("Aadam"))
```

#### Number

```python
## Using function
def palindrome_check(num):
    original = num
    num_sum = 0
    while num>0:
        temp = num%10
        num=num//10
        num_sum = num_sum*10 + temp
    return num_sum == original
        
print(palindrome_check(12321))
```

#### Generate squares using list comprehension

```python
square = [x**2 for x in range(11)]
print(square)
```

#### ARMSTRONG NUMBER CHECKER

```python
def arms_num(num):
    s = str(num)
    p = len(s)
    return sum(int(i)**p for i in s) == num
print(arms_num(15))
```

#### FIZZBUZZ IMPLEMENTATION

```python
for i in range(1,101):
    if i%3 == 0 and i%5 ==0:
        print("FizzBuzz")
    elif i%3==0:
        print("Fizz")
    elif i%5 == 0:
        print("Buzz")
    else:
        print(i)
```
