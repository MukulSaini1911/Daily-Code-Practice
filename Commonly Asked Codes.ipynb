```python
## Reverse a string
## Method 1
str1 = 'Mukul Saini is an AI/ML Engineer'
print(str1[::-1])
```

```python
## Reverse a string
## Method 2
text = "Mukul Saini is an AI/ML Engineer"
reversed_text = ""

for char in text:
    reversed_text = char + reversed_text

print(reversed_text)
```

```python
## find duplicate characters in a string
str2 = "Mukul Saini is a good person"
dup = []
lst = []
for i in str2:
    if i not in dup and str2.count(i)>1:
        dup.append(i)
    else:
        lst.append(i)
print(dup)
```

```python
## find duplicate items in a list
lst2 = ['apple','banana','guava','apple','mango','mango']
dup = []
lst = []
for i in lst2:
    if i not in dup and lst2.count(i)>1:
        dup.append(i)
    else:
        lst.append(i)
print(dup)
```

```python
## find duplicate characters in a list
## Method2

from collections import Counter

fruits = ["apple", "banana", "apple", "cherry", "banana", "apple"]
fruit_counts = Counter(fruits)

print(fruit_counts)           # Output: Counter({'apple': 3, 'banana': 2, 'cherry': 1})
print(fruit_counts["apple"])   # Output: 3

for fruit, count in fruit_counts.items():
    if count > 1:
        print(fruit)
```

```python
## Palindrome string

str3 = '1234321'
if str3 == str3[::-1]:
    print('It is a palindrom')
else:
    print('It is not a palindrom')
    
```

```python
## Palindrome number
n = int(input("Enter the number"))
rev = 0
temp = n
while n > 0:
    digit = n%10
    rev = (rev*10) + digit 
    n = n//10
if rev == temp:
    print("Given number is palindrom.")
else:
    print("Number is not palindrom")
    
```

```python
##find maximum value in list python without inbuilt function
def find_maximum(numbers):
    # Check if the list is empty to prevent an IndexError
    if not numbers:
        return None
        
    # Assume the first element is the largest initially
    max_value = numbers[0]
    
    # Iterate through the rest of the list
    for num in numbers:
        # Update max_value if a larger number is found
        if num > max_value:
            max_value = num
            
    return max_value

# Example usage:
my_list = [10, 24, 76, 23, 12]
print(find_maximum(my_list))  # Output: 76
```

```python
##-----------------------------------
## Binary Search
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2  # Find the middle index
        if arr[mid] == target:
            return mid           # Found it! Return the index
        elif arr[mid] < target:
            low = mid + 1        # Target is in the right half
        else:
            high = mid - 1       # Target is in the left half
            
    return -1                    # Target is not in the list

# Test the function
numbers = [3, 5, 9, 11, 15, 21, 28]
print(binary_search(numbers, 15))  # Output: 4
```

```python
## List comprehension

print([x*2 for x in range(10)]) #Square

print([x for x in range(1,10) if x%2!=0]) #Odd

print([x for x in range(1,10) if x%2==0]) #Even
```

```python
# Decorators
def logger(func):
    def wrapper():
        print(f"Executing {func.__name__}")
        func()
    return wrapper
@logger
def fetch_data():
    print("Fetching data from database")
fetch_data()
```

### OOP Concepts

```python
## Inheritance

class Animal:
    def eat(self):
        print("Animal is eating")

class Dog(Animal):   # Dog inherits from Animal
    def bark(self):
        print("Dog is barking")

d = Dog()

d.eat()    # Inherited method
d.bark()   # Own method
```

```python
## Polimorphism
# Example 1: Method Overriding (Runtime Polymorphism)

class Parent:
    def show(self):
        pass

class Child(Parent):
    def show(self):
        print("Child Method")

c = Child()
c.show()
```

```python
# Example 2: Built-in Polymorphism
print(len("Hello"))
print(len([1, 2, 3, 4]))
```
